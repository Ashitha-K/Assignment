1. An Institution wishes to find out their student’s ability in
mathematics, reading and writing skills. The Institution
wants to do an exploratory study to check the following
information.
a. Find out how many males and females participated in
the test.
b. What do you think about the students' parental level of
education?

```python
# a. To Find out how many males and females participated in the test
gender= data['gender'].value_counts()
data['gender'].value_counts().plot(kind='bar',color="Green")
print(gender)
```

b. What do you think about the students' parental level of
education?

```python
# b. Analyze the students' parental level of education
parental_education= data['parental level of education'].value_counts()
data['parental level of education'].value_counts().plot(kind='pie',)
print(parental_education)
```

```python
# Calculate average scores by parental education level
average_scores_by_education = data.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
average_scores_by_education
plt.plot(average_scores_by_education)
```

c. Who scores the most on average for math, reading and
writing based on
i. Gender
ii. Test preparation course

```python

average_scores_by_gender = data.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
average_scores_by_test_preparation = data.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
print("Average Scores by Gender:\n", average_scores_by_gender)
print("\nAverage Scores by Test Preparation Course:\n", average_scores_by_test_preparation)
```

What do you think about the scoring variation for
math, reading and writing based on
i. Gender
ii. Test preparation course

```python
 #i. Scoring variation by gender
variation_by_gender = data.groupby('gender')[['math score', 'reading score', 'writing score']].agg(['mean', 'std'])

# ii. Scoring variation by test preparation course
variation_by_test_preparation = data.groupby('test preparation course')[['math score', 'reading score', 'writing score']].agg(['mean', 'std'])

print("Scoring Variation by Gender:\n", variation_by_gender)
print("\nScoring Variation by Test Preparation Course:\n", variation_by_test_preparation)
```

e. The management needs your help to give bonus points
to the top 25% of students based on their math score,
so how will you help the management to achieve this.

Case Study on Testing of Hypothesis
2. A company started to invest in digital marketing as a new
way of their product promotions. For that they collected
data and decided to carry out a study on it.
a. The company wishes to clarify whether there is any
increase in sales after stepping into digital marketing.
b. The company needs to check whether there is any
dependency between the features “Region” and
“Manager”.

```python
from scipy import stats
#H0:There is no significance increase in sales after digital marketing
#Ha:There is a significance increase in sales after digital marketing

# Perform paired t-test
t_stat, p_value = stats.ttest_rel(df['Sales_before_digital_add(in $)'],df['Sales_After_digital_add(in $)'],alternative='greater')

t_stat, p_value
```

```python
alpha=0.05
if p_value<alpha:
  print("Reject the null hypothesis:There is a significance increase in sales after digital marketing")
else:
  print("Fail to reject the null hypothesis:There is no significance increase in sales after digital marketing")

```

b. The company needs to check whether there is any dependency between the features “Region” and “Manager”.

```python

from scipy.stats import chi2_contingency


# Example: Test the association between 'Region' and 'Manager'
# Create a contingency table
contingency_table = pd.crosstab(df['Region'], df['Manager'])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-square statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)

# Interpretation
if p < 0.05:
    print("There is a significant association between 'Region' and 'Manager'.")
else:
    print("There is no significant association between 'Region' and 'Manager'.")
```
